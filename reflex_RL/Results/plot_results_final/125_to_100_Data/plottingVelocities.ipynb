```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
import scipy.signal as signal
```

```python
# =========LOAD baselien data===============
file_path = 'baselineVelocities.csv'  # Replace with your file path
data = pd.read_csv(file_path)

baseline_velocities = data.iloc[1:3000, 1]  # Second column (Step)
baseline_velocities_filt = data.iloc[1:3000, 2]  # Third column (Value)
x_range = [x * 0.01 for x in range(len(baseline_velocities))]

# =========LOAD PPO data===============
file_path = 'PPOVelocities.csv'  # Replace with your file path
data = pd.read_csv(file_path)

PPO_velocities = data.iloc[:, 1]  # Second column (Step)
PPO_velocities_filt = data.iloc[:, 2]  # Third column (Value)

x_range_PPO = [x * 0.01*3 for x in range(len(PPO_velocities))]

# =========LOAD PPO data===============
file_path = 'A2CVelocities.csv'  # Replace with your file path
data = pd.read_csv(file_path)

A2C_velocities = data.iloc[:, 1]  # Second column (Step)
A2C_velocities_filt = data.iloc[:, 2]  # Third column (Value)

x_range_A2C = [x * 0.01*3 for x in range(len(A2C_velocities))]

```

```python
# Step 4: Plot the original, noisy, and filtered signals
plt.figure(figsize=(6.5, 4.4))
plt.plot(x_range, baseline_velocities, linestyle='-', color='k', alpha = 0.3, linewidth=0.7)
plt.plot(x_range, baseline_velocities_filt, linestyle='-', color='k', label='CMA-ES', linewidth=4)
plt.plot(x_range_PPO, PPO_velocities, linestyle='-', color='b', alpha = 0.3, linewidth=0.7)
plt.plot(x_range_PPO, PPO_velocities_filt, linestyle='-', color='b', label='PPO', linewidth=4)
plt.plot(x_range_A2C, A2C_velocities, linestyle='-', color='r', alpha = 0.3, linewidth=0.7)
plt.plot(x_range_A2C, A2C_velocities_filt, linestyle='-', color='r', label='A2C', linewidth=4)
plt.legend()
# plt.title("Measured and Filtered Pelvis Velocities", fontsize=16)
plt.xlabel("Simulation Time (s)", fontsize=16)
plt.ylabel("Velocities (m/s)", fontsize=16)
plt.ylim(0.5, 2.0)
plt.grid(True)
plt.savefig('Velocities')
plt.show()
```
